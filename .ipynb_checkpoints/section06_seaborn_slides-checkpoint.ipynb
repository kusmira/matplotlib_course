```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv("Diamonds Prices2022.csv")
```

```python
diamonds.head()
```

```python
sns.barplot(x="cut",
            y="carat",
            data=diamonds)
```

```python
sns.barplot(x="carat",
            y="cut",
            data=diamonds)
```

```python
sns.barplot(x="carat",
            y="cut",
            hue="color",
            data=diamonds.sort_values(by="color"),
            palette="viridis")
```

```python
sns.histplot(x="price", data=diamonds)
```

```python
sns.boxplot(x="price", data=diamonds)
```

```python
sns.boxplot(x="cut",
            y="price",
            data=diamonds,)
```

```python
sns.violinplot(x="color", y="price", data=diamonds)
```

# LMPLOT & REGPLOT

```python
sns.regplot(x="carat", y="price", data=diamonds)
```

```python
sns.lmplot(x="carat", 
           y="price", 
           row="cut",
           col="clarity",
           fit_reg=False,
           #scatter=False,
           hue="cut",
           #by="cut",
           #alpha=.3
           data=diamonds)
```

```python
sns.pairplot(diamonds)
```

```python
sns.heatmap((diamonds
             .pivot_table(index="cut",
                          columns="clarity",
                          values={"carat":"mean"}
                         )
             .droplevel(0, axis=1)
            )
)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})

sns.heatmap((diamonds
             .pivot_table(index="cut",
                          columns="color",
                          values={"price":"mean"}
                         )
             .droplevel(0, axis=1)),
            annot=True,
            fmt='g',
)
```

```python
import numpy as np

sns.set_style("white")

corr = diamonds.drop("Unnamed: 0", axis=1).corr()

# Getting the Upper Triangle of the co-relation matrix
matrix = np.triu(corr)


sns.heatmap(corr, cmap="vlag", mask=matrix) #vmin, #vmax, 
```

```python
sns.jointplot(x="carat", y="price", hue="clarity", data=diamonds)
```

```python
import matplotlib.pyplot as plt

def annotate(data, **kws):
    n = len(data)
    ax = plt.gca()
    ax.text(.7, .1, f"N = {n}", transform=ax.transAxes)

g = sns.FacetGrid(diamonds, col="color", col_wrap=3)
g.map_dataframe(sns.scatterplot, x="carat", y="price")
g.map_dataframe(annotate)
```

```python
# Combo w/ matplotlib

fig, ax = plt.subplots(2, 2)

sns.regplot(ax=ax[0][0], x="carat", y="price", data=diamonds)

sns.heatmap((diamonds
             .pivot_table(index="cut",
                      columns="clarity",
                      values={"price":"mean"}
                     )
             .droplevel(0, axis=1)
        ),
            ax = ax[0][1]
)

ax[1][0].hist(diamonds["price"])

ax[1][1].bar(x=diamonds["color"], height=diamonds["price"])

plt.show()
```
