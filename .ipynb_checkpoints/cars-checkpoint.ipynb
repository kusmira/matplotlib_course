```python
top10 = list(cars.Manufacturer.value_counts().iloc[:10].index)

top10
```

```python
avg_price = cars.groupby("Manufacturer").agg({"Price": "mean"})

avg_price = avg_price.loc[avg_price.index.isin(top10), "Price"].reindex(top10)

avg_price
```

```python
cars.Manufacturer.value_counts().iloc[:10]
```

```python
fig, ax = plt.subplots()

cars = cars.assign(over_120k = cars["Price"] > 120000)

ax.bar(top10, 
       height = cars.Manufacturer.value_counts().iloc[:10],
       width=.3)

ax2 = ax.twinx()

ax2.bar(top10,
        height=avg_price.values,
        width=.3,
      )
```
