```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_raw = pd.read_csv("housing_prices.tsv000", 
                      sep="\t", 
                      parse_dates=["period_begin", "period_end"],
                      dtype=({"total_homes_sold": "Int32"}),
                      usecols=["period_begin", 
                               "period_end", 
                               "region_type",
                               "region_name", 
                               "active_listings",
                               "duration", 
                               "months_of_supply",
                               "total_homes_sold", 
                               "median_active_list_price",
                               "inventory",
                               "age_of_inventory"
                              ]
)
```

```python
ca_housing = housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    & (housing_raw["duration"] == "4 weeks")
    & (housing_raw["region_type"] == "county")
].assign(region_name = housing_raw["region_name"].str[0:-11]).query("region_name == 'San Francisco'")
```

```python
ca_housing.info(memory_usage="deep")
```

```python
ca_housing.head()
```

```python
sf_housing_prices = (ca_housing
                     .set_index("period_begin")
                     .loc[:, "median_active_list_price"]
                     .sort_index()
                     .resample("Q")
                     .mean()
                    )
```

```python
sf_housing_prices.astype("int")
```

```python
fig, ax = plt.subplots()

ax.plot(sf_housing_prices.index, sf_housing_prices)
```

```python
fig, ax = plt.subplots()

ax.plot(sf_housing_prices.index, sf_housing_prices)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.set_title("Housing Prices in San Francisco 2017-2022", fontsize=16)
ax.set_xlabel("Year", fontsize=16)
ax.set_ylabel("Price (Millions)", fontsize=16)


plt.axvline(18352, c="r", ls="--")


ax.annotate("Start of Covid",
            xy=(18300, 1250000),
            xytext=(17600, 1250000),
            arrowprops=dict(facecolor="black", 
                            width=1, 
                            headwidth=8,
                            connectionstyle="angle3, angleA=270, angleB=0"),
            verticalalignment="center"
           )


ax.plot(sf_housing_prices)

plt.show()
```

```python
ca_housing = housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    & (housing_raw["duration"] == "4 weeks")
    & (housing_raw["region_type"] == "county")
].assign(region_name = housing_raw["region_name"].str[0:-11]).query("region_name in ['San Francisco', 'Los Angeles']")
```

```python
ca_housing_markets = (ca_housing
                     .set_index("period_begin")
                     .loc[:, ["region_name", "median_active_list_price"]]
                     .sort_index()
                     .pivot_table(index="period_begin",
                                  columns="region_name",
                                  values="median_active_list_price")
                     .resample("Q")
                     .mean()
                    )
```

```python
ca_housing_markets
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.set_title("Housing Prices in CA Markets 2017-2022", fontsize=16)
ax.set_xlabel("Year", fontsize=16)
ax.set_ylabel("Price (Millions)", fontsize=16)

ax.plot(ca_housing_markets)

plt.legend(ca_housing_markets.columns)

plt.axvline(18352, c="black", ls="--")


ax.annotate("Start of Covid",
            xy=(18300, 1250000),
            xytext=(17600, 1250000),
            arrowprops=dict(facecolor="black", 
                            width=1, 
                            headwidth=8,
                            connectionstyle="angle3, angleA=270, angleB=0"),
            verticalalignment="center"
           )

plt.show()
```

```python
sf_dual = (ca_housing
          .loc[ca_housing["region_name"] == "San Francisco", 
              ["period_begin", "inventory", "median_active_list_price"]]
          .set_index("period_begin")
          .sort_index()
    )
```

```python
fig, ax = plt.subplots()

fig.suptitle("Surge in Inventory Depresses Prices")

ax.plot(sf_dual.index, 
        sf_dual["median_active_list_price"], 
        label="Price",
        c="orange")

ax.set_ylabel("Price (Millions)")

ax2 = ax.twinx()

ax2.plot(sf_dual.index, 
         sf_dual["inventory"], 
         label="Inventory",
         c="green")

ax2.set_ylabel("Inventory")

plt.axvline(18341, c="black", ls="--")

ax.annotate("Start of Covid",
            xy=(18300, 1500000),
            xytext=(17300, 1500000),
            arrowprops=dict(facecolor="black", 
                            width=1, 
                            headwidth=8,
                            connectionstyle="angle3, angleA=270, angleB=0"),
            verticalalignment="center"
           )

fig.legend(bbox_to_anchor=(.9, .88))
```
