```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# list(pd.read_csv("housing_prices.tsv000", 
#                       sep="\t",
#                       nrows=1000).columns)
```

```python
housing_raw = pd.read_csv("housing_prices.tsv000", 
                      sep="\t", 
                      parse_dates=["period_begin", "period_end"],
                      dtype=({"total_homes_sold": "Int32"}),
                      usecols=["period_begin", 
                               "period_end", 
                               "region_type",
                               "region_name", 
                               "active_listings",
                               "duration", 
                               "months_of_supply",
                               "total_homes_sold", 
                               "median_active_list_price",
                               "inventory",
                               "age_of_inventory"
                              ]
)
```

```python
ca_housing = housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    & (housing_raw["duration"] == "12 weeks")
    & (housing_raw["region_type"] == "county")
].assign(region_name = housing_raw["region_name"].str[0:-11])
```

```python
ca_housing.info(memory_usage="deep")
```

```python
ca_housing.head()
```

```python
ca_housing_pivot = ca_housing.pivot_table(index="period_begin",
                                          columns = "region_name",
                                          values = "median_active_list_price").assign(
                                          CA_average = lambda x: x.mean(axis=1))
```

```python
ca_housing_pivot
```

```python
ca_housing.groupby("region_name").agg({"median_active_list_price": "mean"}).sort_values("median_active_list_price", ascending=False)
```

```python
ca_housing_pivot.iloc[-1].mean()
```

```python
ca_housing_markets = ca_housing_pivot.loc[:, ["San Francisco", 
                                              "Los Angeles", 
                                              "San Diego", 
                                              "CA_average"]]
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(ca_housing_markets)

plt.legend(ca_housing_markets.columns)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.set_title("Housing Prices in CA Markets 2017-2022", fontsize=16)
ax.set_xlabel("Year", fontsize=16)
ax.set_ylabel("Price (Millions)", fontsize=16)


ax.plot(ca_housing_markets)

plt.legend(ca_housing_markets.columns)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.set_title("San Francisco Prices Decline After Covid Start", fontsize=16)
ax.set_xlabel("Year", fontsize=16)
ax.set_ylabel("Price (Millions)", fontsize=16)

colors = ["Orange", "Grey", "Grey", "Grey"]

for i in range(len(colors)):
    ax.plot(ca_housing_markets.iloc[:, i], color=colors[i])

plt.legend(ca_housing_markets.columns)

plt.axvline(18341, c="black", ls="--")

ax.annotate("Start of Covid",
            xy=(18300, 1300000),
            xytext=(17600, 1200000),
            arrowprops=dict(facecolor="black", 
                            width=1, 
                            headwidth=8,
                            connectionstyle="angle3, angleA=270, angleB=0"),
            verticalalignment="center"
           )


plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_title("Median List Price CA Cities June 2022", fontsize=12)
ax.set_xlabel("Market", fontsize=12)
ax.set_ylabel("Price (Millions)", fontsize=12)


ax.bar(ca_housing_markets.iloc[-1, :3].index, ca_housing_markets.iloc[-1, :3])
```

```python
avg = ca_housing_pivot.iloc[-1].mean()
median = ca_housing_pivot.iloc[-1].median()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))


ax.set_title("Median List Price CA Cities June 2022", fontsize=12)
ax.set_xlabel("Market", fontsize=12)
ax.set_ylabel("Price (Millions)", fontsize=12)

colors = ["orange", "grey", "grey"]

ax.bar(["SF", "LA", "SD"], ca_housing_markets.iloc[-1, :3], color=colors, label="Median Price")

ax.axhline(median, 0, 1, color="black", ls='dotted', label="California Median")

fig.legend(fontsize=14, bbox_to_anchor=(.9, .88))
```

```python
sales_totals = ca_housing.query("region_name in ['San Francisco', 'Los Angeles', 'San Diego']").groupby("region_name").agg({"total_homes_sold":"sum"})
```

```python
fig, ax = plt.subplots()

ax.pie(x=sales_totals.iloc[:, 0],
       startangle=90,
       labels=["Los Angeles", "San Diego", "San Francisco"], autopct="%.0f%%")


ax.set_title("Share of Home Sales Select CA Markets")
```

```python
price_by_region = housing_raw.groupby(["region_name"]).agg({"median_active_list_price": "mean"}).sort_values("median_active_list_price", ascending=False)
```
