```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
housing_raw = pd.read_csv("housing_prices.tsv000", 
                      sep="\t", 
                      parse_dates=["period_begin", "period_end"],
                      # dtype=({"total_homes_sold": "Int32"}),
                      usecols=["period_begin", 
                               "period_end", 
                               "region_type",
                               "region_name", 
                               "active_listings",
                               "duration", 
                               "months_of_supply",
                               "total_homes_sold", 
                               "median_active_list_price",
                               "median_active_list_ppsf_yoy",
                               "inventory",
                               "age_of_inventory"
                              ]
)
```

```python
ca_housing = (
    housing_raw.loc[
        (housing_raw["region_name"].str.contains("CA"))
        & (housing_raw["duration"] == "4 weeks")
        & (housing_raw["region_type"] == "county")
    ]
    .assign(region_name = housing_raw["region_name"].str[0:-11])
    .query("region_name in ['San Francisco', 'Los Angeles', 'San Diego', 'Tulare']")
    .set_index(["period_begin"])
    .sort_index()
)

ca_housing.head()
```

```python
sf_housing = ca_housing.loc[ca_housing["region_name"] == "San Francisco"]
la_housing = ca_housing.loc[ca_housing["region_name"] == "Los Angeles"]
sd_housing = ca_housing.loc[ca_housing["region_name"] == "San Diego"]
tl_housing = ca_housing.loc[ca_housing["region_name"] == "Tulare"]
```

```python
fig, ax = plt.subplots(2, 2, sharex='all', figsize=(10, 10))

ax[0][0].plot(sf_housing.index, sf_housing["median_active_list_price"])
ax[0][0].set_title("San Francisco")

ax[0][1].plot(sd_housing.index, sd_housing["median_active_list_price"])
ax[0][1].set_title("San Diego")

ax[1][0].plot(tl_housing.index, tl_housing["median_active_list_price"])
ax[1][0].set_title("Tulare")

ax[1][1].plot(la_housing.index, la_housing["median_active_list_price"])
ax[1][1].set_title("Los Angeles")

plt.show()
```
