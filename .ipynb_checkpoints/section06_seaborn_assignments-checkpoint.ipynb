```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hotels = pd.read_excel("HotelCustomersDataset.xlsx").drop(["NameHash", "DocIDHash"], axis=1)

hotels.head()
```

```python
hotels.iloc[:, :12].describe()
```

```python
sns.histplot(x="RoomNights", data=hotels)
```

```python
sns.violinplot(
    x="Nationality",
    y="LodgingRevenue",
    data=hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP']"))
```

```python
sns.violinplot(
    x="Nationality",
    y="LodgingRevenue",
    data=hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP'] \
                        and LodgingRevenue < 1500"))
```

```python
hotels["TotalRevenue"] = hotels["LodgingRevenue"] + hotels["OtherRevenue"]
```

```python
sns.violinplot(
    x="Nationality",
    y="TotalRevenue",
    data=hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP'] \
                        and TotalRevenue < 2000"))
```

```python
sns.violinplot(
    x="Nationality",
    y="Age",
    data=hotels.query("Nationality in ['GBR', 'PRT', 'FRA', 'DEU', 'ESP']"))
```

```python
sns.violinplot(y="RoomNights", data=hotels.query("RoomNights < 15"))
```

```python
sns.regplot(x="RoomNights",
            y="LodgingRevenue",
            data=hotels)
```
