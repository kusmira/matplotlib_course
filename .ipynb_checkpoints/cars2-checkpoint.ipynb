```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv("car_prices.csv", on_bad_lines='skip')
```

```python
cars.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

sns.barplot(data=cars,
            x = "make",
            y = "sellingprice")
```

```python
cars["make"].value_counts().nlargest(15) 
```

```python
cars2 = cars.assign(make2 = cars["make"].where(cars["make"].isin(cars["make"].value_counts().iloc[:15].index), "Other"))

cars2["make2"].value_counts()
```

```python
cars_plot = cars2.loc[cars2["body"]=="sedan"].groupby(["make2"]).agg({"sellingprice": "mean"}).sort_values("sellingprice", ascending=False)
```

```python
sns.barplot(data=cars_plot,
            x = cars_plot.index,
            y = "sellingprice"           
           )
```

```python
cars_agg = cars2.groupby("year").agg(avg_price = ("sellingprice", "mean"), 
                           listing_count = ("year", "count"))
```

```python
fig, ax = plt.subplots()

ax.plot(cars_agg.index, cars_agg["avg_price"], label="Average Price", c="g")

ax2 = ax.twinx()

ax2.plot(cars_agg.index, cars_agg["listing_count"], label="Number of Cars Auctioned")

fig.legend()

## mult color labels
```

```python
fig, ax = plt.subplots(figsize=(10, 10))



bins = np.linspace(1, 5, 9)
labels = ["1-1.5", "1.5-2", "2-2.5", "2.5-3", "3-3.5", "3.5-4", "4-4.5", "4.5-5"]

cars2["condition_bins"] = pd.cut(cars2["condition"], bins=bins, labels=labels)

sns.barplot(x="condition_bins",
            y="sellingprice",
            data=cars2)
```

```python
cars2["condition"].describe()
```

```python
top10 = cars2["make"].value_counts().iloc[:10].index
```

```python
avgs = cars2.loc[cars2["make"].isin(top10)].pivot_table(index="make", columns="condition_bins", values={"sellingprice": "mean"})
```

```python
fig, ax = plt.subplots()

plt.title("Avg. Selling Price by Make and Quality", fontsize=20)

sns.heatmap(avgs,
            annot=True,
            cmap="RdYlGn",
            fmt="g")
```
