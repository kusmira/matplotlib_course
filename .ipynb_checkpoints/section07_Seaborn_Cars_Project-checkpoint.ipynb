```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("car_prices.csv", on_bad_lines='skip')
```

```python
cars.head()
```

# Numeric Variable Relationships

```python
sns.pairplot(cars)
```

```python
sns.heatmap(cars.corr(), cmap="RdYlGn", annot=True)
```

# Categorical Variable Relationships

```python
fig, ax = plt.subplots(figsize=(10, 10))



bins = np.linspace(1, 5, 9)
labels = ["1-1.5", "1.5-2", "2-2.5", "2.5-3", "3-3.5", "3.5-4", "4-4.5", "4.5-5"]

cars["condition_bins"] = pd.cut(cars["condition"], bins=bins, labels=labels)

sns.barplot(x="condition_bins",
            y="sellingprice",
            data=cars)
```

```python
top10 = cars["make"].value_counts().iloc[:10].index

avgs = (cars
        .loc[cars["make"].isin(top10)]
        .pivot_table(index="make", 
                     columns="condition_bins", 
                     values={"sellingprice": "mean"}
                    )
        .droplevel(0, axis=1)
       )

avgs.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))


plt.title("Avg. Selling Price by Make and Quality", fontsize=20)

sns.heatmap(avgs,
            annot=True,
            cmap="RdYlGn",
            fmt="g")
```

```python
f150_pivot = f150s.loc[f150s["year"].between(2009, 2014)].pivot_table(index="trim",
                  columns="condition_bins",
                  values={"sellingprice": "mean"})
```

# Are there any "Deals"

```python
sns.set_style("darkgrid")

f150s["diff_to_mmr"] = f150s["sellingprice"] - f150s["mmr"]

f150s.groupby("color").agg({"diff_to_mmr":"mean"}).plot.bar()
```

```python
f150s_reduced = f150s.query("color != 'orange' and trim != 'SVT Raptor'")

f150r_pivot = f150s_reduced.loc[f150s_reduced["year"].between(2009, 2014)].pivot_table(index="trim",
                  columns="condition_bins",
                  values={"sellingprice": "mean"})

sns.heatmap(f150r_pivot, annot=True, fmt="g")
```
