```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_raw = pd.read_csv("housing_prices.tsv000", 
                      sep="\t", 
                      parse_dates=["period_begin", "period_end"],
                      # dtype=({"total_homes_sold": "Int32"}),
                      usecols=["period_begin", 
                               "period_end", 
                               "region_type",
                               "region_name", 
                               "active_listings",
                               "duration", 
                               
                               "months_of_supply",
                               "total_homes_sold", 
                               "median_active_list_price",
                               "median_active_list_ppsf_yoy",
                               "inventory",
                               "age_of_inventory"
                              ]
)
```

```python
ca_housing = housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    & (housing_raw["duration"] == "4 weeks")
    & (housing_raw["region_type"] == "county")
].assign(region_name = housing_raw["region_name"].str[0:-11]).query("region_name == 'San Francisco'")
```

```python
ca_housing.info(memory_usage="deep")
```

```python
ca_housing.head()
```

```python
sf_housing_prices = (ca_housing
                     .set_index("period_begin")
                     .loc[:, "median_active_list_price"]
                     .sort_index()
                     .resample("Q")
                     .mean()
                    )
```

```python
sf_housing_prices
```

```python
fig, ax = plt.subplots()

ax.plot(sf_housing_prices.index, sf_housing_prices)
```

# Grouped Bar Plot - Note the Pandas API/Seaborn are much easier.

```python
ca_or = (housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    | (housing_raw["region_name"].str.contains("OR"))
    & (housing_raw["duration"] == "4 weeks")
    & (housing_raw["region_type"] == "county")].assign(region_name = housing_raw["region_name"].str[0:-11]))# .query("region_name in ['San Francisco', 'Los Angeles', 'San Diego']")
```

```python
ca_or["state"] = np.where(ca_or["region_name"].str.contains("CA"), "CA", "OR")
```

```python
ca_or = ca_or.groupby([ca_or["state"], ca_or["period_end"].dt.year]).agg({"median_active_list_price": "mean"}).reset_index()
```

```python
ca_or
```

```python
ca_or = ca_or.pivot_table(index="period_end",
                          columns="state",
                          values={"median_active_list_price": "mean"}).droplevel(0, axis=1)

ca_or
```

```python
fig, ax = plt.subplots()

width=.35
x=np.arange(6)

bar1= ax.bar(x-width/2, ca_or["CA"], width=width)
bar2 = ax.bar(x+width/2, ca_or["OR"], width=width, color="orange")

plt.xticks(x, ca_or.index, fontsize=8)

plt.legend([bar1, bar2], ["California", "Texas"])
```

```python
ca_yearly = (housing_raw.loc[
    (housing_raw["region_name"].str.contains("CA"))
    & (housing_raw["duration"] == "4 weeks")
    & (housing_raw["region_type"] == "county")].assign(region_name = housing_raw["region_name"].str[0:-11])).query("region_name in ['San Francisco', 'Los Angeles', 'San Diego']")

ca_yearly.head()
```

```python
ca_yearly = (ca_yearly.groupby([ca_yearly["region_name"], ca_yearly["period_end"].dt.year])
             .agg({"total_homes_sold": "sum"})
             .reset_index())

ca_yearly = ca_yearly.pivot_table(index="period_end",
                          columns="region_name",
                          values={"total_homes_sold": "count"}).droplevel(0, axis=1)

ca_yearly
```

```python
fig, ax = plt.subplots()

x = np.arange(0, len(ca_yearly))
#width = 0.35

ax.bar(ca_yearly.index, ca_yearly["Los Angeles"], label="Los Angeles")
ax.bar(ca_yearly.index, ca_yearly["San Diego"], label="San Diego", bottom=ca_yearly["Los Angeles"]) 


ax.set_title("Revenue Country and Category")
ax.set_ylabel("Revenue")
ax.set_xlabel("Country")
ax.legend(bbox_to_anchor=(1,1))

fig.tight_layout()
```

```python
ca_yearly_pct = ca_yearly.apply(lambda x: x * 100 / sum(x), axis=1)

ca_yearly_pct
```

```python
fig, ax = plt.subplots()

x = np.arange(0, len(ca_yearly))
#width = 0.35

# Initialize the bottom at zero for the first set of bars.
bottom = np.zeros(len(ca_yearly))


for i, col in enumerate(ca_yearly_pct.columns):
  ax.bar(ca_yearly_pct.index, ca_yearly_pct[col], bottom=bottom, label=col)
  bottom += np.array(ca_yearly_pct[col])

ax.set_title("Revenue Country and Category")
ax.set_ylabel("Revenue")
ax.set_xlabel("Country")
ax.legend(bbox_to_anchor=(1,1
                         ))

#plt.xticks(range(4), ca_yearly.index)

fig.tight_layout()
```
